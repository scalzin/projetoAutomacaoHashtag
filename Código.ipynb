### Passo 1 - Importar Arquivos e Bibliotecas

```python
import pandas as pd
import pathlib 
import win32com.client as win32

caminho = (r'C:\Users\Luana\Documents\One Drive Pessoal\OneDrive\Cursos\HASHTAG\Projetos\Projeto 1\Bases de Dados')
email = pd.read_excel(f'{caminho}\Emails.xlsx')
vendas = pd.read_excel(f'{caminho}\Vendas.xlsx')
lojas = pd.read_csv(f'{caminho}\Lojas.csv', encoding='latin-1', sep=';')

display(lojas)
display(email)
display(vendas)
```

### Passo 2 - Definir Criar uma Tabela para cada Loja e Definir o dia do Indicador

```python
vendas = vendas.merge(lojas, on='ID Loja')
display(vendas)
```

```python
dicLojas = {}
for loja in lojas['Loja']:
    dicLojas[loja] = vendas.loc[vendas['Loja'] == loja, :]
print(dicLojas['Shopping Barra'])
```

```python
dia = vendas['Data'].max()
print('{}/{}'.format(dia.day, dia.month))
```

### Passo 3 - Salvar a planilha na pasta de backup

```python
caminhoBK = pathlib.Path(r'Backup Arquivos Lojas')
pastaBK = caminhoBK.iterdir()

nomesBK = []
for arquivo in pastaBK:
    nomesBK.append(arquivo.name)
print(nomesBK)
```

```python
for loja in dicLojas:
    try:
        if loja not in nomesBK:
            novaPasta = caminhoBK / loja
            novaPasta.mkdir()
    except FileExistsError :
        pass
    
    nomeArquivo = '{}_{} {}.xlsx'.format(dia.day, dia.month, loja)
    localArquivo = caminhoBK / loja / nomeArquivo
    
    dicLojas[loja].to_excel(localArquivo)
```

### Passo 4 - Calcular o indicador para 1 loja

```python
#metas
faturamentoAno = 1650000
faturamentoDia = 1000
produtosAno = 120
produtosDia = 4
ticketAno = 500
ticketDia = 500
```

```python
for shop in dicLojas:
    vendasLojas = dicLojas[shop]
    vendasLojasDia = vendasLojas.loc[vendasLojas['Data']==dia, :]

    #faturamento
    fatuAno = vendasLojas['Valor Final'].sum()
    fatuDia = vendasLojasDia['Valor Final'].sum()

    #diversidade dos produtos
    prodAno = len(vendasLojas['Produto'].unique())
    prodDia = len(vendasLojasDia['Produto'].unique())

    #ticket medio
    valorVenda = vendasLojas.groupby('Código Venda').sum(numeric_only=True)
    tktAno = valorVenda['Valor Final'].mean()
    valorVendaDia = vendasLojasDia.groupby('Código Venda').sum(numeric_only=True)
    tktDia = valorVendaDia['Valor Final'].mean()
    
    #email
    outlook = win32.Dispatch('outlook.application')

    gerente = email.loc[email['Loja'] == shop , 'Gerente'].values[0]
    mail = outlook.CreateItem(0) 
    mail.To = email.loc[email['Loja'] == shop , 'E-mail'].values[0]
    mail.Subject = 'OnePage dia {}/{} - Loja {}'.format(dia.day, dia.month, shop)

    if fatuDia >= faturamentoDia:
        corFatuDia = 'green'
    else:
        corFatuDia = 'red'

    if fatuAno >= faturamentoAno:
        corFatuAno = 'green'
    else:
        corFatuAno = 'red'

    if prodDia >= produtosDia:
        corProdDia = 'green'
    else:
        corProdDia = 'red'

    if prodAno >= produtosAno:
        corProdAno = 'green'
    else:
        corProdAno = 'red'

    if tktDia >= ticketDia:
        corTktDia = 'green'
    else:
        corTktDia = 'red'

    if tktAno >= ticketAno:
        corTktAno = 'green'
    else:
        corTktAno = 'red'

    mail.HTMLBody = f'''
    <p> Bom dia, {gerente}!</p>

    <p> O resultado de ontem <strong>(dia {dia.day}/{dia.month})</strong> da <strong>Loja {shop}</strong> foi:</p>

    <html>
    <head>

    </head>
    <body>

    <table>
      <tr>
        <th>Indicador</th>
        <th>Valor Dia</th>
        <th>Meta Dia</th>
        <th>Cenário Dia</th>
      </tr>
      <tr>
        <td>Faturamento</td>
        <td style='text-align:center'>R$ {fatuDia:.2f}</td>
        <td style='text-align:center'>R$ {faturamentoDia:.2f}</td>
        <td style='text-align:center'><font color ="{corFatuDia}">◙</font></td>
      </tr>
      <tr>
        <td>Diversidade Produtos</td>
        <td style='text-align:center'>{prodDia}</td>
        <td style='text-align:center'>{produtosDia}</td>
        <td style='text-align:center'><font color ="{corProdDia}">◙</font></td>
      </tr>
      <tr>
        <td>Ticket Médio</td>
        <td style='text-align:center'>R$ {tktDia:.2f}</td>
        <td style='text-align:center'>R$ {ticketDia:.2f}</td>
        <td style='text-align:center'><font color ="{corTktDia}">◙</font></td>
      </tr>
      <tr>
    </table>

    <br>

    <table>
      <tr>
        <th>Indicador</th>
        <th>Valor Ano</th>
        <th>Meta Ano</th>
        <th>Cenário Ano</th>
      </tr>
      <tr>
        <td>Faturamento</td>
        <td style='text-align:center'>R$ {fatuAno:.2f}</td>
        <td style='text-align:center'>R$R$ {faturamentoAno:.2f}</td>
        <td style='text-align:center'><font color ="{corFatuAno}">◙</font></td>
      </tr>
      <tr>
        <td>Diversidade Produtos</td>
        <td style='text-align:center'>{prodAno}</td>
        <td style='text-align:center'>{produtosAno}</td>
        <td style='text-align:center'><font color ="{corProdAno}">◙</font></td>
      </tr>
      <tr>
        <td>Ticket Médio</td>
        <td style='text-align:center'>R$ {tktAno:.2f}</td>
        <td style='text-align:center'>R$ {ticketAno:.2f}</td>
        <td style='text-align:center'><font color ="{corTktAno}">◙</font></td>
      </tr>
      <tr>
    </table>


    </body>
    </html>


    <p>Segue em anexo a planilha com todos os dados para mais detalhes. </p>

    <p>Qualquer dúvida estou a disposição </p>
    <p> Atenciosamente, Luana</p>
    '''

    anexo = pathlib.Path.cwd() / caminhoBK / shop / f'{dia.day}_{dia.month} {shop}.xlsx'
    mail.Attachments.Add(str(anexo))

    mail.Send()

    #th -> titulo
    #td -> cada valor dentro da tabela
    #tr -> linha da tabela
    print(f'Email da Loja {shop} enviado')
```

### Passo 7 - Criar ranking para diretoria

```python
fatuLojas = vendas.groupby('Loja')[['Loja', 'Valor Final']].sum(numeric_only=True)
fatuLojasAno = fatuLojas.sort_values(by='Valor Final', ascending=False)
display(fatuLojasAno)

nomeArquivo = '{}_{} Ranking Anual.xlsx'.format(dia.day, dia.month)
fatuLojasAno.to_excel(r'Backup Arquivos Lojas\{}'.format(nomeArquivo))

vendasDia = vendas.loc[vendas['Data'] == dia, :]
fatuLojasDia = vendasDia.groupby ('Loja')[['Loja', 'Valor Final']].sum(numeric_only=True)
fatuLojasDia = fatuLojasDia.sort_values(by='Valor Final', ascending=False)
display(fatuLojasDia)

nomeArquivoDia = '{}_{} Ranking Diário.xlsx'.format(dia.day, dia.month)
fatuLojasDia.to_excel(r'Backup Arquivos Lojas\{}'.format(nomeArquivoDia))
```

### Passo 8 - Enviar e-mail para diretoria

```python
outlook = win32.Dispatch('outlook.application')

mail = outlook.CreateItem(0) 
mail.To = email.loc[email['Loja'] == 'Diretoria' , 'E-mail'].values[0]
mail.Subject = 'Ranking Diário {}/{}'.format(dia.day, dia.month)
mail.Body = f'''
Prezados, bom dia!

Melhor loja do dia em Faturamento: Loja {fatuLojasDia.index[0]} com Faturamento R$ {fatuLojasDia.iloc[0,0]:.2f}
Pior loja do dia em Faturamento: Loja {fatuLojasDia.index[-1]} com Faturamento R$ {fatuLojasDia.iloc[-1,0]:.2f}

Melhor loja do ano em Faturamento: Loja {fatuLojasAno.index[0]} com Faturamento R$ {fatuLojasAno.iloc[0,0]:.2f}
Pior loja do ano em Faturamento: Loja {fatuLojasAno.index[-1]} com Faturamento R$ {fatuLojasAno.iloc[-1,0]:.2f}

Segue em anexo os rankings do ano e do dia de todas as lojas.

Qualquer dúvida estou a disposição.

Atenciosamente, 
Luana
'''

anexo = pathlib.Path.cwd() / caminhoBK / f'{dia.day}_{dia.month} Ranking Anual.xlsx'
mail.Attachments.Add(str(anexo))
anexoDia = pathlib.Path.cwd() / caminhoBK / f'{dia.day}_{dia.month} Ranking Diário.xlsx'
mail.Attachments.Add(str(anexoDia))

mail.Send()

#th -> titulo
#td -> cada valor dentro da tabela
#tr -> linha da tabela
print('Email da Loja Diretoria enviado')
```
